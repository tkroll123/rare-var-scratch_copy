```python
# build a graph with potentially random order of the entries
def build_graph(g, graph_str, shuffle = False):
    # create a list with possible graph entries
    graph_entries = []
    for i in graph_str.splitlines():
        if len(i)>0:
            # check if more than one edge can be reached from the starting edge;
            neighbor_str = i.split('[')[1].split(']')[0]
            neighbor_list = neighbor_str.split(',')
            n_neighbors = len(neighbor_list)
            if n_neighbors>1:
                for j in range(n_neighbors):
                    graph_entries.append(i.split('[')[0] + '[' + neighbor_list[j].replace(' ','') + '])')
            else:
                # delete all empty sapces
                graph_entries.append(i.replace(' ',''))

    # randomly reorder the graph entries
    if shuffle == True:
        import random
        random.shuffle(graph_entries)
    
    # interpret the string in list to build the graph
    for i in graph_entries:
        exec(i)

```

```python
# ---------------------------------------------------------------   
from termcolor import colored
import sys

# ---------------------------------------------------------------   
# Helper functions:
class operations():
        
    def get_suffix(self, value):
        tmp = str(value).split(".")
        return(int(tmp[1]))
        
    def get_prefix(self, value):
        return(int(value))

    
# ---------------------------------------------------------------        
# This class represents a directed graph using adjacency list representation
class DFS_graph():
 
    # Constructor
    def __init__(self):
        self.graph = {}
        self.tr_nodes = []

        self.verbose = False
        
        #op = operations()

        self.counter = 0
        self.shortPaths = {}
        self.shortPathNodes = {}
        self.edgeMarkers = {}
        self.nodeMarkers = {}
        self.upstream_graph = {}
        self.active_nodes = set()
        self.active_edges = set()
        self.dead_nodes = set()
        self.dead_edges = set()
        
    def reinitialize(self):
        self.counter = 0
        self.shortPaths = {}
        self.shortPathNodes = {}
    
    # add edges to the graph dictionary:
    def addEdge(self, edge, neighbors):
        if edge not in self.graph.keys():
            self.graph[edge] = neighbors
        else:
            for neighbor in neighbors:
                if neighbor not in self.graph[edge]:
                    self.graph[edge].append(neighbor)
                    
        # create upstream dictionary where the values are the edges from which an edge can be reached.
        for neighbor in neighbors:
            if neighbor not in self.upstream_graph.keys():
                self.upstream_graph[neighbor] = [edge]
            else:
                self.upstream_graph[neighbor].append(edge)

        # mark edge as unvisited
        self.edgeMarkers[edge] = 'not visited'
            
    # add entries to the dictionary that contains short paths without a fork:
    def addShortPath(self, edge, node):
        if self.counter not in self.shortPaths.keys():
            self.shortPaths[self.counter] = [edge]
            self.shortPathNodes[self.counter] = [node]
        else:
            self.shortPaths[self.counter].append(edge)
            self.shortPathNodes[self.counter].append(node)

            
    # erase the dead branch from the shortPaths dictionary and set the edges to 'dead'
    def prune(self):
        # set the whole current dictionary branch to dead in the edgehMarkers dictionary,
        for edge in self.shortPaths[self.counter]:
            self.edgeMarkers[edge] = 'dead'
            self.dead_edges.add(edge)
        for node in self.shortPathNodes[self.counter]:
            self.nodeMarkers[node] = 'dead'
            self.dead_nodes.add(node)
        # delete that branch in the shortPaths dictionary
        del self.shortPaths[self.counter]
        del self.shortPathNodes[self.counter]
       
        # check if only one path out of the previous edge is possible after one dangling path got removed:
        if self.counter > 0:
            last_edge = self.shortPaths[self.counter-1][-1]
            outgoing_edge = last_edge
            outgoing_edge_markers = []
            for i in self.graph[outgoing_edge]:
                outgoing_edge_markers.append(self.edgeMarkers[i])
            if self.verbose:
                print('outgoing_edge:', outgoing_edge)
                print('outgoing_edge_markers:', outgoing_edge_markers)
            all_dead = (outgoing_edge_markers.count('dead') == len(outgoing_edge_markers) & len(outgoing_edge_markers)>0 )
            if (all_dead):
                if self.verbose:
                    print(colored('{} has only dead edges. Recursive loop in self.prune.'.format(last_edge), 'red'))
                self.counter -= 1
                self.prune()

    # set all edges and nodes to active within a given list (i.e. short paths)
    def set_path_active(self, edges, nodes):
        for i in edges:
            self.edgeMarkers[i] = 'active'
            self.active_edges.add(i)
        for i in nodes:
            self.nodeMarkers[i] = 'active'
            self.active_nodes.add(i)

    # combine slected paths segments to one bigger one and
    # rename the ones that are left so that the shortPaths.keys() are in order
    def combine_paths_segments(self, list_in_shortPath, list_out_shortPath):
        if self.verbose:
            print(colored('combine paths segments (new):', 'magenta'))
            print("list_in_shortPath:", list_in_shortPath)
        
        for n in range(1,len(list_in_shortPath)):
            counter = list_in_shortPath[n]
            # compine the edges:
            for m in range(len(self.shortPaths[counter])):
                if self.shortPaths[counter][m] not in self.shortPaths[list_in_shortPath[0]]:
                    self.shortPaths[list_in_shortPath[0]].append(self.shortPaths[counter][m])
            del self.shortPaths[counter]
            # compine the nodes:
            for m in range(len(self.shortPathNodes[counter])):
                if self.shortPathNodes[counter][m] not in self.shortPathNodes[list_in_shortPath[0]]:
                    self.shortPathNodes[list_in_shortPath[0]].append(self.shortPathNodes[counter][m])
            del self.shortPathNodes[counter]

        if self.verbose:
            print("done combining path segments!!")
            print('before self.shortPaths.keys() got cleaned:')
            print('new short paths:', self.shortPaths)
            print('new short nodes:', self.shortPathNodes)
        
        counter = list_in_shortPath[0]
        for i in list_out_shortPath:
            counter = counter +1
            if i != counter:
                self.shortPathNodes[counter] = self.shortPathNodes[i]
                del self.shortPathNodes[i]
                self.shortPaths[counter] = self.shortPaths[i]
                del self.shortPaths[i]
            
        self.counter = counter
        if self.verbose:
            print('after self.shortPaths.keys() got cleaned:')
            print('new short paths:', self.shortPaths)
            print('new short nodes:', self.shortPathNodes)
            print('new counter:', self.counter)
            print('done with def combine_paths_segments')


    # build the list with short paths segments that need and do not need to be combined later
    def select_path_segments(self, list_active_shortPath, list_notactive_shortPath, active_edge, i):
        if self.verbose:
            print("check which nodes are active:")
            print("Active nodes:", self.active_nodes)
        # find upstream neighbors from active edge:
        if (active_edge in self.upstream_graph.keys() and self.counter - i >= 0):
            upstream_neighbors = self.upstream_graph[active_edge]
            # check if any of the upstream neighbors are in the previous short path:
            
            if self.verbose:
                print("set(upstream_neighbors) from edge {}: {}".format(active_edge, set(upstream_neighbors)))
                print("self.counter-i:", self.counter-i)
                print("set(self.shortPaths[self.counter-i]):", set(self.shortPaths[self.counter-i]))
            
            upstream_overlap = set(upstream_neighbors) & set(self.shortPaths[self.counter-i])
            if len(upstream_overlap) >0:
                list_active_shortPath.append(self.counter-i)
                # the new active edge is the first edge in self.shortPaths[self.counter-i]:
                active_edge = self.shortPaths[self.counter-i][0]
            else:
                list_notactive_shortPath.append(self.counter-i)
                
        return list_active_shortPath, list_notactive_shortPath, active_edge

    
    # if an active node or tandem repeat node has been found, combine the short paths that are contained 
    # in the path to one bigger short path
    def combine_active_path(self):
        cnt = 0
        i = 1

        list_active_shortPath = [self.counter]
        list_notactive_shortPath = []

        active_edge = self.shortPaths[self.counter][0]
        
        if self.verbose:
            print('active nodes:', self.active_nodes)
        while True:
            if self.verbose:
                print(" ***** create active paths ******************")
                print('cnt =', cnt)
            if cnt == 4:
                break
            if self.verbose:
                print("i = ", i)
                print('short paths:', self.shortPaths)
                print('short nodes:', self.shortPathNodes)
                print('self.shortPathNodes[self.counter-i]', self.shortPathNodes[self.counter-i])

            list_active_shortPath, list_notactive_shortPath, active_edge = \
            self.select_path_segments(list_active_shortPath, list_notactive_shortPath, active_edge, i)
                    
            overlap = self.active_nodes & set(self.shortPathNodes[self.counter-i])
            if len(overlap)>0:
                if self.verbose:
                    print('active node(s) {} found for i={}'.format(overlap, i))
                # combine all elements that are in the list list_active_shortPath
                self.combine_paths_segments(sorted(list_active_shortPath), sorted(list_notactive_shortPath))
                # set path to active
                if self.verbose:
                    print('active short path segment: =', sorted(list_active_shortPath)[0])
                    
                self.set_path_active(self.shortPaths[sorted(list_active_shortPath)[0]], self.shortPathNodes[sorted(list_active_shortPath)[0]])
                break

            i += 1
            cnt += 1
            if cnt == 5:
                print(colored('break because of cnt limit!!','red'))
                break
            
            if i > self.counter+1:
                print(colored('self.counter dropped below -1! That is impossible! Abort!', 'red'))
                print('i:', i, 'self.counter', self.counter)
                break
    
        if self.verbose:
            print(" ***** done with create active paths ******************")     
  
    # if a loop has been found, combine the short paths that are contained in the loop to 
    # one bigger short path
    def combine_loop(self, loop_node):
        cnt = 0
        i = 1
        loop_counter = -1

        list_active_shortPath = [self.counter]
        list_notactive_shortPath = []

        active_edge = self.shortPaths[self.counter][0]
        
        if self.counter == 0:
            if self.verbose:
                print(" ***** combine loop ******************")
                print('self.counter = 0 !!')
        else:    
            while True:
                if self.verbose:
                    print(" ***** combine loop ******************")
                    print('cnt =', cnt)
                if cnt == -4:
                    break
                if self.verbose:
                    print("Loop node:", loop_node)
                    print("i = ", i)
                    print('short paths:', self.shortPaths)
                    print('short nodes:', self.shortPathNodes)
                    
                list_active_shortPath, list_notactive_shortPath, active_edge = \
                self.select_path_segments(list_active_shortPath,list_notactive_shortPath, active_edge, i)    
                    
                if loop_node in self.shortPathNodes[self.counter-i]:
                    if self.verbose:
                        print('found loop_node {} in element number {}'.format(loop_node, self.counter-i))
                    loop_counter = self.counter-i
                    # combine all elements that are in the list list_active_shortPath
                    self.combine_paths_segments(sorted(list_active_shortPath), sorted(list_notactive_shortPath))
                    break
                else:
                    if self.verbose:
                        print('found loop_node {} not in element number {}'.format(loop_node, self.counter-i))

                i += 1
                cnt += 1
                if cnt == -5:
                    print(colored('break because of cnt limit!!','red'))
                    break

                if i == self.counter + 1:
                    # first check if the loop node is a tandem repeat node:
                    if loop_node in self.shortPathNodes[-1]:
                        if self.verbose:
                            print('loop node found in self.shortPathNodes[-1], which is the starting node in tandem repeat region')
                            print('Combine paths till self.counter = 0')
                        self.combine_paths_segments(sorted(list_active_shortPath), sorted(list_notactive_shortPath))
                    else:
                        if self.verbose:
                            print('The starting node has been reached, which is not the loop node.')
                            print('Hence, this is not a loop, just continue from here.')
                    break

                if i > self.counter:
                    print(colored('self.counter dropped below 0! That is impossible! Abort!', 'red'))
                    print(i, self.counter)
                    sys.exit()

            # After the whole loop is combined, check if the loop node is active. 
            # If so, set the whole loop to active
            if self.nodeMarkers[loop_node]== 'active':
                if self.verbose:
                    print(colored('loop_node {} is active. Set the whole loop to active.'.format(loop_node), 'black'))
                    print(colored('loop_counter: {}'.format(loop_counter), 'black'))                    
                if loop_counter == -1:
                    loop_counter = 0
                self.set_path_active(self.shortPaths[loop_counter], self.shortPathNodes[loop_counter])
                
        if self.verbose:
            print(" ***** end of combine loop ******************")
        
        
    # --------------------------------------------------------------------------------------------
    # A function used by DFS for recursive approach
    def DFSUtil(self, edge, visited_edges, visited_nodes):

        # -- preparatory steps: ----------------------------------------
        if self.verbose:
            print(colored('current edge: {}'.format(edge), 'red'))
        
        # add edge to those that got visited
        visited_edges.add(edge)
                
        # label the current edge as 'undecided'
        self.edgeMarkers[edge] = 'undecided'
        
        # get the node that is reached
        node = op.get_suffix(edge)
        if self.verbose:
            print(colored('current node: {}'.format(node), 'green'))
        
        # mark as undecided if not yet assigned a value:
        if node not in self.nodeMarkers.keys():
            self.nodeMarkers[node] = 'undecided'

        # add to the short paths
        self.addShortPath(edge,node)
        if self.verbose:
            print('self.shortPaths = ', self.shortPaths)
            print('self.shortPathNodes = ', self.shortPathNodes)

        # get all neighbors from the current edge:
        neighbor_edges = self.graph[edge]      # this is a list
        
        # make a set of neighbor character ('alive', 'dead', 'undecided')
        neighbor_markers = set()
        number_of_dead_neighbor_edges = 0
        for i in neighbor_edges:
            if i in self.edgeMarkers.keys():
                neighbor_markers.add(self.edgeMarkers[i])
                if self.edgeMarkers[i]=='dead':
                    number_of_dead_neighbor_edges += 1
            
            
        # -- check for active edges, dangling edges, etc: -------------------------------------

        # check if a node has been visited or not:
        if node in visited_nodes:   # i.e. we looped around
            # combine those edges into one path until a node is reached that has been visited already
            if self.verbose:
                print(colored('*** Already visited node found: {} - Go to self.combine_loop()'.format(node), 'blue'))
            self.combine_loop(node)
        else:
            visited_nodes.add(node)
        
        # Dangling Node is found:
        if len(neighbor_edges) == 0 and op.get_suffix(edge) not in self.tr_nodes:
            if self.verbose:
                print(colored('*** Dangling edge found: {}'.format(edge), 'blue'))
            self.prune()
    
        # check if an edge connects to only dead edges
        if (len(neighbor_markers)==1) and ('dead' in neighbor_markers):
            if self.verbose:
                print(colored('*** Connection only to dead edges found: {}'.format(edge), 'blue'))
            self.prune()
            
        # Connection to Tandem Repeat region is found
        if op._getsuffix(edge) in self.tr_nodes:# and node not in visited_nodes:
            if self.verbose:
                print(colored('*** Tandem repeat node found: {} from edge {}'.format(op.get_suffix(edge), edge), 'blue'))
            node = op.get_suffix(edge)
            self.nodeMarkers[node] = 'alive'
            self.edgeMarkers[edge] = 'alive'
            self.active_nodes.add(node)
            self.combine_active_path()
            
        # check if an edge connects to one alive edge
        if (len(neighbor_markers)>0) and ('alive' in neighbor_markers):# and (node not in visited_nodes):
            if self.verbose:
                print(colored('*** New connected edge found: {}'.format(edge), 'blue'))
            node = op.get_suffix(edge)
            if self.verbose:
                print(colored('    Active node {}'.format(node), 'blue'))
            self.active_nodes.add(node)
            self.combine_active_path()

            
        # -- the recursive part: --------------------------------------------------------------
        if edge in self.graph.keys():
            # check if there are more than one options, or if all options are already traversed 
            # or if we hit a tandem repeat node.
            if ((len(neighbor_edges)-number_of_dead_neighbor_edges >  1) or 
                (set(neighbor_edges).issubset(visited_edges) and len(neighbor_edges)>0) or
                op.get_suffix(edge) in self.tr_nodes):
                if len(self.shortPaths.keys())>0:
                    self.counter = max(self.shortPaths.keys())+1

            for ind, neighbor_edge in enumerate(neighbor_edges):
                if neighbor_edge not in visited_edges:
                    self.DFSUtil(neighbor_edge, visited_edges, visited_nodes)

        
    # The function to do DFS traversal. It uses recursive DFSUtil()
    def DFS(self, edge):

        visited_edges = set()
        visited_nodes = set()
        
        # add the starting node
        starting_node = op.get_prefix(edge)
        visited_nodes.add(starting_node)
        self.nodeMarkers[starting_node] = 'active'
        
        self.active_nodes.add(starting_node)
        self.shortPathNodes[-1] = [starting_node]
        
        # Call the recursive helper function
        # to print DFS traversal
        self.DFSUtil(edge, visited_edges, visited_nodes)
        
        # if only one node is active, that means everything else is dead and the only active 
        # node is the staring node.
        # Set to 'erased'
        if len(self.active_nodes)==1:
            self.active_nodes.remove(starting_node)
            self.nodeMarkers[starting_node]='erased'
            

        if self.verbose:
            print(colored('self.edgeMarkers: {}'.format(self.edgeMarkers), 'red'))
            print(colored('self.nodeMarkers: {}'.format(self.nodeMarkers), 'green'))
        
        
```
